```python
# Mengimport Library yang akan digunakan
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import data

img = cv2.imread("images/badak.jpg", cv2.IMREAD_GRAYSCALE)  # Load gambar dari local drive dalam format grayscale
row, column = img.shape                                     # Mendapatkan dimensi citra
img1 = np.zeros((row,column),dtype = 'uint8')               # Membuat array kosong dengan dimensi yang sama dengan citra
min_range = 10                                              # Menetapkan nilai batas bawah
max_range = 60
 
# Membuat perulangan untuk mengubah piksel menjadi 255 atau 0 berdasarkan batas rentang
for i in range(row):
    for j in range(column):
        if img[i,j]>min_range and img[i,j]<max_range:
            img1[i,j] = 255
        else:
            img1[i,j] = 0
```

```python
# PLOTING IMAGE

fig, axes = plt.subplots(2, 2, figsize=(12, 12))    # Membuat subplot dengan 2 baris dan 2 kolom
ax = axes.ravel()                                   # Mengubah array subplot menjadi 1 dimensi

ax[0].imshow(img, cmap=plt.cm.gray)                 # Menampilkan citra input 
ax[0].set_title("Citra Input")  

ax[1].hist(img.ravel(), bins=256)                   # Menampilkan histogram citra input
ax[1].set_title('Histogram Input') 

ax[2].imshow(img1, cmap=plt.cm.gray)                # Menampilkan hasil citra output 
ax[2].set_title("Citra Output")  

ax[3].hist(img1.ravel(), bins=256)                  # Menampilkan histogram citra output
ax[3].set_title('Histogram Output')  
```
