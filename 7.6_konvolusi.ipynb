```python
# Mengimport Library yang akan digunakan
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
```

```python
citra1 = imread(fname="images/badak.jpg")  # Load image dengan nama file "badak.jpg" menggunakan fungsi imread.
print(citra1.shape)                 # Menampilkan informasi dimensi citra
plt.imshow(citra1, cmap='gray')     # Menampilkan citra menggunakan imshow dengan cmap 'gray'. 
```

```python
# Membuat kernel atau area masking untuk konvolusi berukuran 9x9
kernel = np.array([[-1, 0, -1], 
                   [0, 4, 0], 
                   [-1, 0, -1]])

citraOutput = cv2.filter2D(citra1, -1, kernel)      # Melakukan filtering citra menggunakan kernel dengan filter2D. Parameter (-1) menunjukkan bahwa tipe output citra akan memiliki kedalaman yang sama dengan citra input
fig, axes = plt.subplots(1, 2, figsize=(12, 12))    # Membuat objek subplot dengan ukuran 1x2 dan ukuran gambar 12x12.
ax = axes.ravel()                                   # Melakukan ravel pada objek axes untuk mendapatkan array dari semua sumbu subplot.

# Menampilkan citra input  & output pada subplot
ax[0].imshow(citra1, cmap='gray') 
ax[0].set_title("Citra Input") 
ax[1].imshow(citraOutput, cmap='gray') 
ax[1].set_title("Citra Output") 
```
